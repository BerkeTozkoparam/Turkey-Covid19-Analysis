# Turkey COVID-19 Pandas Analysis

In this project, Turkey's COVID-19 process was analyzed with data from Our World in Data.

## Scope
- Daily case and death charts
- Basic statistics (total cases, deaths, testing)
- Highest days of cases/deaths

## Libraries Used
- pandas
- matplotlib

```python
import pandas as pd

#Read data skipping line errors
df = pd.read_csv("compact.csv", on_bad_lines='skip')

# Write the new clean file 
df.to_csv("compact_clean.csv", index=False)
```

```python
df.head()
```

# Values of Turkey's Covid19

```python
turkiye = df[df["country"] == "Turkey"]




```

```python
print(turkiye)
```

# Days of maximum deaths and cases

```python
print(turkiye.sort_values("new_cases", ascending=False).head(1))
print(turkiye.sort_values("new_deaths", ascending=False).head(1))
```

# Total statistic

```python
print("Total_case:", turkiye["new_cases"].sum())
print("Total_deaths:", turkiye["new_deaths"].sum())
print("Total_tests:", turkiye["new_tests"].sum())
```

# Daily case/mortality graphs (matplotlib)
